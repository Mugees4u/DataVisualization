```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Style
   'linestyle' or 'ls' used for change the style of the plotted line

```python
#Example: Use Dotted Line

y = np.array([6, 7, 9, 3])

plt.plot(y, linestyle='dotted')

z = np.array([5, 8, 11, 6])

plt.plot(z, ls='dashed')
plt.show()
```

# Line Color

    'color' or 'c'

```python
t = [16, 12, 18]
plt.plot(t, c='r')

z = [7, 22, 54]
plt.plot(z, ls='--', color='g')

plt.show()
```

# Line Width

          'linewidth'
    
           'lw'

```python
y = [13, 19, 27, 8]

z = [69, 15, 62]

plt.plot(y, c='r', lw='15.8')

plt.plot(z, color='g', linestyle="dashed", linewidth='20.5')
```
