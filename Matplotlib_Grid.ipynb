# Specify Which Grid Lines to Display

 We can use axis parameter in grid() to specify which grid line to display
    
    

```python
x = [56, 58, 69, 63]
y = [54, 59, 70, 69]

plt.plot(x, y)

plt.grid(axis='x')  # Grid the line for only x axis

plt.show()
```

```python
x = [56, 58, 69, 63]
y = [54, 59, 70, 69]

plt.plot(x, y)

plt.grid(axis='y')   # Grid the line for the y axis only

plt.show()
```

```python
x = [56, 58, 69, 63, 93]
y = [54, 59, 70, 69, 86]

f3 = {'family':'serif', 'color':"green" , 'size':20}
plt.plot(x, y)

plt.title('Grid Line with all the properties', fontdict=f3)

plt.grid(c='r', ls='dotted', lw='1')

plt.show()
```
