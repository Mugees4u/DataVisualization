```python
import numpy as np
import matplotlib.pyplot as plt
```

# Create a Labels for a Plot

        With Pyplot we can use the xlabel()  and ylabel() to set the labels for x and y as wel

```python
x = [56, 14, 23, 25]

y = [25000, 28000, 32000, 12300]

plt.plot(x, y)

plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

# Create a Title for the Plot 

```python
x = [3, 10, 12, 5]
y = [13500, 67000, 87000, 42000]

f1 = {'family':'serif', 'color':'blue', 'size':20}
f2 = {'family':'serif', 'color':'red', 'size':20}
f3 = {'family':'serif', 'color':'green', 'size':22}

plt.plot(x, y, c='r', ls='dashed', lw='2.5')
plt.xlabel('Working Experince', fontdict=f1)
plt.ylabel('Monthly Income', fontdict=f2)
plt.title('Employe Salary Structure', fontdict=f3, loc='left')
plt.show()
```
