```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Display Multiple Plots

   with 'subplot()', can draw multiple plots in one figure

```python
#plot 1

x = [56, 14, 22, 23]
y = [58, 16, 24, 26]
plt.subplot(2, 1, 1)
plt.plot(x,y, c='r')

#plot 2

x = [46, 24, 32, 33]
y = [48, 26, 34, 36]
plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1

x = [55, 13, 21, 22]
y = [57, 15, 23, 25]
plt.subplot(1, 2, 1)
plt.plot(x,y, c='g')
plt.title('First Plot')

#plot 2

y = [49, 27, 33, 34]
x = [47, 25, 35, 37]
plt.subplot(1, 2, 2)
plt.title('Second Plot')

plt.plot(x,y, c='r')

plt.suptitle('This is main Title')
plt.show()
```
